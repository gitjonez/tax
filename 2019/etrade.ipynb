```python
import pandas as pd
from datetime import date
```

```python
BASE = '/Users/dennis/Documents/Google Drive/Etrade/'
et = pd.read_csv(BASE + 'DownloadTxnHistory2019.csv',
        skiprows=1, 
        parse_dates=['TransactionDate']
    ).sort_values(
        by='TransactionDate',
        ascending=True
    )
et.to_csv(f'mahle-etrade-2019-transactions.csv', index=False)
et['TransactionDate']
```

```python
et.columns
```

```python
byType = {}
for xtype in pd.unique(et['TransactionType']):
    print(xtype)
    df = et[et['TransactionType'] == xtype].sort_values(by='TransactionDate', ascending=True)
    byType[xtype] = df
    if xtype in ['Dividend', 'Interest', 'Fee']
        df.to_csv(f'mahle-etrade-2019-{xtype}.csv', index=False)
```

```python
trades = pd.concat([byType['Bought'], byType['Sold']], ignore_index=True
    ).sort_values(by='TransactionDate', ascending=True)
trades
```

```python
trades.to_csv('mahle-etrade-2019-trades.csv', index=False)
```
