```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
date_arr = np.arange('2020-01-25', '2020-02-15', dtype='datetime64[D]')
```

```python
df = pd.DataFrame({'Day': date_arr, 'Value': np.zeros(len(date_arr), dtype= 'int')})
```

```python
df
```

```python
data = [
    (date(2020, 1, 26), 3),
    (date(2020, 1, 27), 2),
    (date(2020, 1, 29), 4)
]
```

```python
data
```

```python
data[0][0]
```

```python
data_df = pd.DataFrame(data, columns=['Day', 'Value'])
data_df['Day'] = pd.to_datetime(data_df['Day'])
data_df
```

```python
pd.concat([df, data_df], join='inner').sort_values('Day').groupby('Day').sum().reset_index()
```
